```python
# prompt: pesquisar os top 10 países com imigrantes do canada.
top_10=df.sort_values(by='Total', ascending=False).head(10)
```

# Seaborn tem temas:


```python
sns.set_theme()
```

```python
ax=sns.barplot(data=top_10,y=top_10.index,x='Total', orient='h')
ax.set(
    title='Os dez países com mais imigrantes para o Canadá',
    xlabel='Total de imigrantes',
    ylabel=''
)
# com ax.set, consigo passar vários valores em um DataFrame
plt.show()
```

```python
# Criando uma função para lidar com geração de gráficos
def gerar_grafico_palete(palete):
   fig,ax=plt.subplots(figsize=(8,4))
   ax=sns.barplot(data=top_10,y=top_10.index,x='Total', orient='h', palette=palete)
   ax.set_title('Os países com mais imigrantes para o Canadá', loc='left',fontsize=16)
   ax.set_xlabel('Total de imigrantes', fontsize=14)
   ax.set_ylabel('')
   plt.show()
# Assim se inicia uma função
```

```python
gerar_grafico_palete("mako")
```

```python
# Utilizando a função
sns.set_theme(style='ticks')
gerar_grafico_palete("mako")
```

```python
# Utilizando a função
sns.set_theme(style='darkgrid')
gerar_grafico_palete("mako")
```
